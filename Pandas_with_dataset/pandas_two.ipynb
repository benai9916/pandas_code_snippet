```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/survey_results_public.csv')
```

## Modifying row and columns data

```python
df.head(2)
```

## changing column name

```python
# will change the column name in the df_copy dataframe
df_copy = df.copy(deep=True)
```

```python
df_copy.rename(columns={'respondent': 'indexNo'}, inplace=True)
df_copy.head(1)
```

```python
# make change in all the column name, have to iterate over the column names

df_copy.columns = [x.lower() for x in df.columns]
```

```python
df_copy.columns
```

## change columns values

```python
df_copy.loc[0, 'mainbranch'] = ('I AM STUDENT')
```

```python
df_copy.head(1)
```

```python
# in first row change the column 'mainbranch' value to lower
df_copy['mainbranch'] = df_copy.loc[0, 'mainbranch'].lower()
```

```python
df_copy.head(2)
```

## map()

```python
# change yes to True and No to False
df_copy['hobbyist'] = df_copy['hobbyist'].map({'Yes': True, 'No': False})
```

# cleaning data

## dropna()

```python
# if will drop all the rows that contain null values

df_copy.dropna()
```

```python
# we can pass in the axis to either drop row or columns
# in blow clase all other column have missing values except 'main branch'

df_copy.dropna(axis=1)
```

```python
# if we pass an parameter 'how=all' then it will drop only those row or column where all the values are missing from a row or column

df_copy.dropna(axis=1, how='all')
```

```python
# with paramater 'how=any' it will drop all the rows or column, where even a single values are missing

df_copy.dropna(axis=1, how='any')
```

```python
# to drop some specific column , we can pass subset

# it will drop column hobbyist, if there is a missing values

df_copy.dropna(axis=0, how='any', subset=['hobbyist'])
```

```python
# we can pass in multiple columns

df_copy.dropna(axis=0, how='any', subset=['hobbyist', 'country'])
```

## replace()

```python
# to repalce the custom missing values like 'na' or 'missing' or 'none'

df_copy.replace('na', np.NaN)
```

## isna()

```python
# check if there is NaN values
# if it is NaN value then it will return true

df_copy.isna()
```

## fillna()

```python
# fill the missing value with the desired value

df_copy.fillna(0).tail()
```
