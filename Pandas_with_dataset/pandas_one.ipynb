## Filtering - Using Conditionals to Filter Rows and Columns

## isin()

## use string method to filter

```python
df.loc[filter_language, 'LanguageWorkedWith'].head(5)
```

```python
# dummy example to apply filter
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
people_df = pd.DataFrame(people)
people_df
```

```python
# in column 'last' value == 'Doe'
filter_people = (people_df['last'] == 'Doe')
```

```python
filter_people
```

```python
# print the filtered result
#  'email' is optional

people_df.loc[filter_people, 'email']
```

```python
# another wau to print the filtered result
people_df[filter_people]
```

 ## Logical operator   '&'  '|' 

```python
# filter using & (and) operator
filter_people = (people_df['last'] == 'Doe') &  (people_df['first'] == 'John') 
people_df[filter_people]
```

```python
# filter using | (or) operator
filter_people = (people_df['last'] == 'Schafer') | (people_df['first'] == 'John') 
people_df[filter_people]
```

```python
# filter using ~ (negation) operator, will give the oposite result
filter_people = (people_df['last'] == 'Schafer') | (people_df['first'] == 'John') 
people_df[~filter_people]
```
