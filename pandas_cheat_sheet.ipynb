# Join and Combine

## concat()

```python
first_df = df.iloc[0:3, 4:7]
second_df = df.iloc[3:5, 4:7]
```

```python
# combine both the df as a column

pd.concat([first_df, second_df])
```

## merge()

```python
# we will the two dataframe on "Opensource" it will be an inner join which will return on the matching column by Opensource

pd.merge(first_df,second_df, on='OpenSource', how='inner')
```
